```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_loss_relu = np.loadtxt("trained_models\\May-19ArchGNNReLUVersion2TrainLoss.txt")
test_loss_relu = np.loadtxt("trained_models\\May-19ArchGNNReLUVersion2TestLoss.txt")
train_loss_leaky = np.loadtxt("trained_models\\May-19ArchGNNVersion1TrainLoss.txt")
test_loss_leaky = np.loadtxt("trained_models\\May-19ArchGNNVersion1TestLoss.txt")
print(train_loss_relu.shape)
print(test_loss_relu.shape)
print(train_loss_leaky.shape)
print(test_loss_leaky.shape)
```

```python
epochs = np.arange(1, 101)
print(epochs.shape)
```

```python
title = 'Loss data for ReLU vs. LeakyReLU over 100 epochs'
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(3)
plt.plot(epochs, train_loss_relu, label='Training loss ReLU')
plt.plot(epochs, test_loss_relu, 'b--', label='Test loss ReLU')
plt.plot(epochs, train_loss_leaky, label='Training loss LeakyReLU')
plt.plot(epochs, test_loss_leaky, 'r--', label='Test loss LeakyReLU')
plt.xlabel('Epochs [#n]')
plt.ylabel('Loss [MSE]')
plt.yscale('log')
plt.grid()
plt.legend()
plt.title(title)
plt.show()
```
