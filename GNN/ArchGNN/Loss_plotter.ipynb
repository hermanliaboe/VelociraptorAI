```python
leaky_durations = []

# Open the file
with open('trained_models\\leaky_relu.txt', 'r') as file:
    # Loop over each line in the file
    for line in file:
        # Split the line into components based on comma
        components = line.split(',')
        # The second component has the duration
        duration_component = components[1]
        # Split the duration component based on ':' and strip whitespace
        duration = duration_component.split(':')[1].strip()
        # Remove the trailing 's' from duration and convert to float
        duration = float(duration[:-1])
        # Add to the list
        leaky_durations.append(duration)

# Now durations contains all the durations as float values
print(len(leaky_durations))

relu_durations = []

# Open the file
with open('trained_models\\relu.txt', 'r') as file:
    # Loop over each line in the file
    for line in file:
        # Split the line into components based on comma
        components = line.split(',')
        # The second component has the duration
        duration_component = components[1]
        # Split the duration component based on ':' and strip whitespace
        duration = duration_component.split(':')[1].strip()
        # Remove the trailing 's' from duration and convert to float
        duration = float(duration[:-1])
        # Add to the list
        relu_durations.append(duration)

# Now durations contains all the durations as float values
print(len(relu_durations))
```

```python
title = 'Time duration of epochs for ReLU vs. LeakyReLU over 100 epochs'
plt.style.use('scientific')
f = plt.figure()

plt.plot(epochs1, relu_durations, label='ReLU')
plt.plot(epochs1, leaky_durations, label='LeakyReLU')

plt.xlim(0,100)
#plt.ylim(0,50)
plt.xlabel('Epochs [#n]')
plt.ylabel('Time [s]')
#plt.yscale('log')
plt.grid()
plt.legend()
plt.title(title)
plt.show()
```
