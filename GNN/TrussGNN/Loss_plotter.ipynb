```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_leaky = np.loadtxt("trained_models\\May-25TrussGNNleaky2TrainLoss.txt")
test_leaky = np.loadtxt("trained_models\\May-25TrussGNNleaky2TestLoss.txt")

train_relu = np.loadtxt("trained_models\\May-25TrussGNNrelu3TrainLoss.txt")
test_relu = np.loadtxt("trained_models\\May-25TrussGNNrelu3TestLoss.txt")
```

```python
epochs = np.arange(1, len(train_leaky)+1)
```

```python
plt.style.use('scientific')
title = 'Training loss plotted over 100 epochs'
f = plt.figure()
fig, axes = plt.subplots(1, 2)

# plt.plot(epochs, train_relu,  label='Training loss ReLU')
axes[0].plot(epochs, train_leaky)
axes[0].text(len(train_leaky)+1.05, train_leaky[-1], f"{train_leaky[-1]:.4f}", ha='left', va='center', color='#4C72B0')

axes[0].set_xlim(0, 100)
axes[0].set_xlabel('Epochs [#n]')
axes[0].set_ylabel('Loss [L1]')
axes[0].grid()
axes[0].set_title(title)


title = 'Test loss plotted over 100 epochs'
axes[1].plot(epochs, test_leaky, color='#55A868')
axes[1].text(len(test_leaky)+1.05, test_leaky[-1], f"{test_leaky[-1]:.4f}", ha='left', va='center', color='#55A868')

axes[1].set_xlim(0, 100)
axes[1].set_xlabel('Epochs [#n]')
axes[1].set_ylabel('Loss [L1]')
axes[1].grid()
axes[1].set_title(title)

plt.subplots_adjust(wspace=0.4, hspace=0.2)  # Increase the spacing between subplots


#plt.legend()
plt.show()
```

```python
train = np.loadtxt("trained_models\\Jun-01_1TrussGNN_train_early_final.txt")
test = np.loadtxt("trained_models\\Jun-01_1TrussGNN_test_early_final.txt")

train_best = np.loadtxt("trained_models\\Jun-01_1TrussGNN_train_early.txt")
test_best = np.loadtxt("trained_models\\Jun-01_1TrussGNN_test_early.txt")

under_835 = []
under_835_i = []
best = 835
for i in range(0, len(test)):
    if test[i] < best:
        under_835.append(test[i])
        under_835_i.append(i)
        best = test[i]
epochs1 = np.arange(1, len(test)+1)
```

```python
title = 'TrussNN Training Plot with Early Saving Implemented'
plt.style.use('scientific')
f = plt.figure()
plt.plot(epochs1, train, color='#4C72B0', label='training')
plt.text(len(train)+ 1.05, train[-1] + 0.3, f"{train[-1]:.2f}", ha='left', va='center', color='#4C72B0')
#plt.scatter(epochs1[under_835_i], train[under_835_i], color='black', label='Early saved models', zorder = 10)
#plt.scatter(epochs1[len(train_best)], train_best[-1], color='red', label='Train loss for best model = '+str(round(train_best[-1], 2)), zorder = 20)

plt.plot(epochs1, test, color='#55A868', label='test')
plt.text(len(test)+ 1.05, test[-1] - 0.3, f"{test[-1]:.2f}", ha='left', va='center', color='#55A868')
plt.scatter(epochs1[under_835_i], test[under_835_i], color='black', label='Early saved models', zorder = 10)
plt.scatter(epochs1[len(test_best)], test_best[-1], color='red', label='Best test loss = '+str(round(test_best[-1],2)), zorder = 20)

plt.xlim(0,1000)
#plt.ylim(0,50)
plt.xlabel('Epochs [#n]')
plt.ylabel('Loss [L1]')

plt.grid()
plt.legend()
plt.title(title)
plt.show()

title = 'Test loss with early saving'
plt.style.use('scientific')
f = plt.figure()

plt.plot(epochs1, test, color='#55A868', label='test')

plt.text(len(test)+ 1.05, test[-1], f"{test[-1]:.2f}", ha='left', va='center', color='#55A868')
plt.scatter(epochs1[under_835_i], test[under_835_i], color='black', label='Early saved models', zorder = 10)
plt.scatter(epochs1[len(test_best)], test_best[-1], color='red', label='Test loss for best model = '+str(round(test_best[-1],2)), zorder = 20)

plt.xlim(0,1000)
#plt.ylim(0,50)
plt.xlabel('Epochs [#n]')
plt.ylabel('L1 Loss')

plt.grid()
plt.legend()
plt.title(title)
plt.show()
```
